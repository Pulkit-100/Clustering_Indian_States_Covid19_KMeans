```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_india = pd.read_csv ("https://api.covid19india.org/csv/latest/state_wise.csv")
```

```python
df_india
```

```python
df = df_india[["State", "Confirmed","Recovered","Deaths"]]
```

```python
df = df.drop(index = 0)
```

```python
df
```

```python
for i in df.iterrows():
    if i[1]["Confirmed"]:        
        recoverPercent = (i[1]['Recovered']/i[1]["Confirmed"])*100
        deathPercent = (i[1]['Deaths']/i[1]["Confirmed"])*100
        df.at[i[0],'Recovered'] = recoverPercent
        df.at[i[0],'Deaths'] =  deathPercent
```

```python
df.columns = ["State", "Confirmed", "Recovered %", "Death %"]
```

```python
df
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(df['Confirmed'],df['Recovered %'],df['Death %'])
ax.set_xlabel('Confirmed Cases')
ax.set_ylabel('Recovered %')
ax.set_zlabel('Deaths %')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Confirmed','Recovered %','Death %']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
# Using k = 4 as per elbow method
```

```python
km1 = KMeans(n_clusters=4, random_state = 1)
y_predicted1 = km1.fit_predict(df[['Confirmed','Recovered %','Death %']])
y_predicted1
```

```python
y_predicted1
```

```python
df["Zone"] = y_predicted1
```

```python
pk = []
for i in range(4):
    pk.append((i,df[df["Zone"]==i]["Confirmed"].sum()/len(df[df["Zone"]==i])))
pk.sort(reverse = True,key = lambda a:a[1])
mapper = {}
for j,i in enumerate(pk):
    mapper[j] = i[0]
mapper
```

```python
df = df.replace({"Zone":mapper})
```

```python
df
```

```python
try:
    df.at[df[df["State"]=="Odisha"].index,'State'] = "Orissa"
except:
    pass
```

```python
df_1 = df[df.Zone==0]
df_2 = df[df.Zone==1]
df_3 = df[df.Zone==2]
df_4 = df[df.Zone==3]

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(df_1['Confirmed'],df_1['Recovered %'],df_1['Death %'], color = "green",label="Cluster1")
ax.scatter(df_2['Confirmed'],df_2['Recovered %'],df_2['Death %'], color = "orange",label="Cluster2")
ax.scatter(df_3['Confirmed'],df_3['Recovered %'],df_3['Death %'], color = "red",label="Cluster3")
ax.scatter(df_4['Confirmed'],df_4['Recovered %'],df_4['Death %'], color = "black",label="Cluster4")
ax.set_xlabel('Confirmed Cases')
ax.set_ylabel('Recovered %')
ax.set_zlabel('Deaths %')

```

```python

import matplotlib.cm

from mpl_toolkits.basemap import Basemap

from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize

import difflib
```

```python
fig, ax = plt.subplots() 
m = Basemap(resolution='c', projection='merc', lat_0=54.5, lon_0=-4.36, llcrnrlon=68., llcrnrlat=6., urcrnrlon=97., urcrnrlat=37.)

m.drawmapboundary(fill_color='#46bcec')
m.fillcontinents(color='#f2f2f2', lake_color='#46bcec')
m.drawcoastlines()

m.readshapefile('India_SHP/INDIA','INDIA')


satlist = []

states = set()

for state_info in m.INDIA_info:
    state = state_info['ST_NAME']
    sat1 = 0
    states.add(state)
    for x in df.iterrows():
        if (x[1]["State"].upper() == state.upper()) or (x[1]["State"].upper() in state.upper()):
            sat1 = x[1]["Zone"]+1
            break
    satlist.append(sat1)

    
df_poly = pd.DataFrame({'shapes':[Polygon(np.array(shape), True) for shape in m.INDIA],
                       'area':[area['ST_NAME'] for area in m.INDIA_info],
                       'satlist': satlist})

shapes = [Polygon(np.array(shape), True) for shape in m.INDIA]
cmap = plt.get_cmap('Reds')

pc = PatchCollection(shapes, zorder=2)

norm = Normalize()
pc.set_facecolor(cmap(norm(df_poly['satlist'].fillna(0).values)))
ax.add_collection(pc)

mapper = matplotlib.cm.ScalarMappable(cmap=cmap)
mapper.set_array(satlist)
# plt.colorbar(mapper, shrink=0.4)

ax.set_title('Covid-19 Zones')
plt.rcParams['figure.figsize'] = [15,15]
plt.show()
```
